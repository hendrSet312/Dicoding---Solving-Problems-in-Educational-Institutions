```python
#membersihkan kolom course
def clean_course(x):
    if x in [33, 9003,9119,9130]:
        return 'Science & Technology'
    elif x in [171,9070,9773, 9853,9238,8014]:
        return 'Social Science'
    elif x in [9147,9670,9991,9254]:
        return 'Business & management'
    return 'Health Science'

#membersihkan kolom country
def clean_country(x):
     if x in [101,109,108,41]: 
         return 'Latin America'
     elif x in [105,103,100,62]:
         return 'East Europe'
     elif x in [26,25,26,24,22,21]:
         return 'Africa'
     elif x in [14,17,2,13]:
         return 'North & West Europe'
     return 'South europe'

#membersihkan kolom pekerjaan
def clean_occupation(x):
    if x in [1, 112, 114]:
        return 'Management'
    elif x in [2, 121, 122, 123, 124]:
        return 'Professional'
    elif x in [3, 131, 132, 134, 135]:
        return 'Technician'
    elif x in [4, 141, 143, 144]:
        return 'Administrative'
    elif x in [5, 151, 152, 153, 154]:
        return 'Service & sales'
    elif x in [6, 7, 8, 9, 161, 163, 171, 172, 174, 175, 181, 182, 183, 192, 193, 194, 195]:
        return 'Labour'
    elif x in [10, 101, 102, 103]:
        return 'Armed Forces'
    return 'Unemployed'

#membersihkan kolom edukasi
def clean_education(x):
    if x in [4,43]:
        return 'Master'
    elif x in [5,44]:
        return 'Doctorate'
    elif x in [2,3,18,39,40,41,42]:
        return 'Bachelor & specialized education'
    elif x in [1,9,10,12,14,19,27,29]:
        return 'High school'
    elif x in [11,26,30,38]:
        return 'Middle school'
    elif x in [36,37]:
        return 'Primary'
    return 'No education'
```

```python
#membersihkan data sebelum dianalisis
def clean_df(df):
    #feature encoding
    df['Marital_status'] = df['Marital_status'].apply(lambda x : 'single' if x == 1 else ('married' if x in [2,5] else 'separated'))
    df['Course'] = df['Course'].apply(lambda x : clean_course(x))
    df['Nacionality'] = df['Nacionality'].apply(lambda x : clean_country(x))
    df['Gender'] = df['Gender'].apply(lambda x : 'Male' if x == 1 else 'Female')
    df['Daytime_evening_attendance'] = df['Daytime_evening_attendance'].apply(lambda x : 'Daytime' if x == 1 else 'Evening')
    
    for cat in  ['qualification','occupation']:
        for person in ['Previous','Mothers','Fathers']:
            if cat == 'qualification':
                df[f'{person}_{cat}'] = df[f'{person}_{cat}'].apply(lambda x : clean_education(x))
                df = df.rename({f'{person}_{cat}':f'{person}_education'},axis = 1)
            else : 
                if person != 'Previous':
                    df[f'{person}_{cat}'] = df[f'{person}_{cat}'].apply(lambda x : clean_occupation(x))

    #feature engineering
    df['1st_yr_approved_unit'] = (df['Curricular_units_1st_sem_approved'] + df['Curricular_units_2nd_sem_approved'])
    df['1st_yr_enrolled_unit'] = (df['Curricular_units_1st_sem_enrolled'] + df['Curricular_units_2nd_sem_enrolled']) 
    df['1st_yr_evaluated_unit'] = (df['Curricular_units_1st_sem_evaluations'] + df['Curricular_units_2nd_sem_evaluations'])
    df['1st_yr_credited_unit'] = (df['Curricular_units_1st_sem_credited'] + df['Curricular_units_2nd_sem_credited'])
    df['Admission_grade'] = df['Admission_grade'] / 2
    df['Previous_qualification_grade'] = df['Previous_qualification_grade'] / 2

    df['1st_yr_unit_grade'] = df[['Curricular_units_1st_sem_grade','Curricular_units_2nd_sem_grade']].mean(axis = 1)
    df['1st_yr_unit_grade'] = df['1st_yr_unit_grade'] * 5
    
    df['1st_yr_Completion_unit_rate'] =  (df['1st_yr_approved_unit']/df['1st_yr_enrolled_unit']).fillna(0)
    df['1st_yr_Success_evaluation_rate'] = (df['1st_yr_approved_unit']/df['1st_yr_evaluated_unit']).fillna(0)

    #mengubah tipe data kolom menjadi boolean
    for col in ['Debtor','Displaced','Educational_special_needs',
                'Tuition_fees_up_to_date','Scholarship_holder','International']:
        df[col] = df[col].astype('bool')

    #merubah nama kolom 
    df = df.rename(columns = {
        'Nacionality': 'Nationality',
        'Daytime_evening_attendance':'Attendance_time',
    })


    #menghapus kolom 
    df = df.drop(columns = ['GDP','Inflation_rate','Unemployment_rate','Application_mode','Application_order',
                            'Curricular_units_1st_sem_approved','Curricular_units_2nd_sem_approved','Curricular_units_1st_sem_enrolled',
                            'Curricular_units_2nd_sem_enrolled','Curricular_units_1st_sem_evaluations','Curricular_units_2nd_sem_evaluations',
                            'Curricular_units_1st_sem_credited','Curricular_units_2nd_sem_credited','Curricular_units_1st_sem_grade',
                            'Curricular_units_2nd_sem_grade','Curricular_units_2nd_sem_without_evaluations','Curricular_units_1st_sem_without_evaluations'])
    
    return df
df = clean_df(df)
```

```python
#menampilkan tabel rata - rata performa dan umur mahasiswa
def show_perfomrance_table(df):
    target_col = ['1st_yr_Success_evaluation_rate','1st_yr_Completion_unit_rate','1st_yr_unit_grade','Age_at_enrollment']
    return df.groupby('Status').median(numeric_only=True)[target_col]


def show_parent_plot(df,par = 'Mothers'):
    par_name = f"{par}_occupation"
    job_unique = df[par_name].unique().tolist()
    row = len(job_unique)//2
    col = 2

    if len(job_unique) % 2 != 0 : 
        row +=1

    fig,axes = plt.subplots(row,col,figsize = (15,15))

    plt.suptitle(f'Student Status Based on {par} Occupation')

    for i in range(len(job_unique)):
        job_name = job_unique[i]
        job_counts = df[df[par_name] == job_name]['Status'].value_counts()
        labels = job_counts.index.tolist()

        if len(job_counts) < 3 : 
            for lab in ['Dropout','Graduation','Enrolled']:
                if lab not in labels:
                    new_series = pd.Series([0],index = [lab])
                    job_counts = pd.concat([job_counts,new_series])
        job_counts.plot(kind = 'bar', ax = axes[i//2,i%2], title = job_name, color = ['red','green','blue'],width = 0.8)
        axes[i//2,i%2].set_xticklabels(labels = ['Dropout','Graduate','Enrolled'],rotation=0)

    for j in range(len(job_unique), col*row):
        fig.delaxes(axes[j//2][j%2])

    fig.tight_layout()

    
def contingency_table_parent_status(df,par):
    par_occu = pd.crosstab(df[f'{par}_occupation'],df['Status'])
    par_occu = par_occu.div(par_occu.sum(axis = 1),axis = 0)
    return par_occu
    
```

## Data Preparation / Preprocessing

```python
#mengubah data status menjadi numerik 
df['Status'] = df['Status'].apply(lambda x : x == 'Dropout')
```

```python
#memecah dataset menjadi train dan test
from sklearn.model_selection import train_test_split

selected_x_col = ['Course', 'Fathers_occupation', 'Mothers_occupation', 'Debtor', 'Gender','Scholarship_holder', 'Age_at_enrollment', '1st_yr_approved_unit', '1st_yr_enrolled_unit', '1st_yr_evaluated_unit', '1st_yr_unit_grade', '1st_yr_Completion_unit_rate', '1st_yr_Success_evaluation_rate']
x = df.drop(['Status'],axis = 1)[selected_x_col]
y = df['Status']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=20)
```

```python
import category_encoders as ce 

#binary encoding 
def binary_encoding(df):
    df['Gender'] = df['Gender'].apply(lambda x :  x == 'Male')
    return df

#catboost encoding
def catboost_encoding(X_train, X_test, y_train, y_test):
    cb_encoder = ce.cat_boost.CatBoostEncoder() 
    X_train = cb_encoder.fit_transform(X_train,y_train)
    X_test = cb_encoder.transform(X_test)
    return X_train,X_test,cb_encoder

#melakukan feature encoding
def feature_engineering(X_train,y_train,X_test,y_test):
    X_train = binary_encoding(X_train)
    X_test = binary_encoding(X_test)
    X_train,X_test,cb_encoder = catboost_encoding(X_train, X_test, y_train, y_test)
    return X_train,X_test,cb_encoder

X_train,X_test,cb_encoder = feature_engineering(X_train,y_train,X_test,y_test)
    
    
```

```python
#oversampling, karena data status tidak seimbang
from imblearn.over_sampling import ADASYN

smote = ADASYN(random_state=42)
X_train,y_train = smote.fit_resample(X_train,y_train)
```

## Modelling

### Hypertuning Parameter

```python

from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import tqdm as notebook_tqdm
import optuna

def objective(trial,data=[X_train,X_test],target=[y_train, y_test]):
    X_train,X_test = data
    y_train, y_test = target

    param = {
        'lambda': trial.suggest_float('lambda', 1e-3, 10.0),
        'alpha': trial.suggest_float('alpha', 1e-3, 10.0),
        'colsample_bytree': trial.suggest_categorical('colsample_bytree', [0.3,0.4,0.5,0.6,0.7,0.8,0.9, 1.0]),
        'subsample': trial.suggest_categorical('subsample', [0.4,0.5,0.6,0.7,0.8,1.0]),
        'learning_rate': trial.suggest_categorical('learning_rate', [0.008,0.01,0.012,0.014,0.016,0.018, 0.02]),
        'n_estimators': 10000,
        'max_depth': trial.suggest_categorical('max_depth', [5,7,9,11,13,15,17]),
        'random_state': trial.suggest_categorical('random_state', [42]),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 300),
    }

    model = XGBClassifier(**param)  
    
    model.fit(X_train,y_train,eval_set=[(X_test,y_test)],early_stopping_rounds=500,verbose=False)
    
    preds = model.predict(X_test)
    
    acc = accuracy_score(y_test, preds)
    
    return acc
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20)
print('Number of finished trials:', len(study.trials))
print('Best trial:', study.best_trial.params)
```

### Modeling

```python
#modelling memakai model xgboost

model = XGBClassifier(**study.best_trial.params)
model.fit(X_train,y_train,eval_set=[(X_test,y_test)],early_stopping_rounds=500,verbose=False)
```

## Evaluation

```python
#evaluasi model 
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test,y_pred))
```
